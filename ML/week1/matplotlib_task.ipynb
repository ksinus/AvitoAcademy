```python
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(20, 10))
fig.suptitle('Plot examples', fontsize=20)

axes[0, 0].set_title('Scatter plot example')
cov = [[1, 0.7], [0.7, 1]]
mean = [20, 10]
x, y = np.random.multivariate_normal(mean, cov, 5000).T
axes[0, 0].scatter(x, y, alpha=0.3, edgecolors='black')
axes[1, 0].set_ylim([7, 13])
axes[1, 0].set_xlim([6, 25])

axes[0, 1].set_title('Hist plot example')
hist_grid = np.random.geometric(p=0.1, size=3000)
axes[0, 1].hist(hist_grid, bins=40, color='red')

axes[1, 0].set_title('Line plot example')
axes[1, 0].set_ylim([-4, 2])
axes[1, 0].set_xlim([0, 6])
x_grid = np.linspace(0.01, 7, 3000)
axes[1, 0].plot(x_grid, np.log(x_grid), label='log(x)')
axes[1, 0].plot(x_grid, np.log(2 * x_grid), label='log(2x)')
axes[1, 0].legend()

axes[1, 1].set_title('Bar plot example')
date_range = np.arange(np.datetime64('2021-01-01'),
                       np.datetime64('2021-01-14'))
mu, sigma = 5, 1
bar_grid = np.random.normal(mu, sigma, len(date_range))
axes[1, 1].bar(date_range, height=bar_grid, width=0.7)
axes[1, 1].tick_params(axis='x', rotation=90)
```
